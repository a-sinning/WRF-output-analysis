```python
#modifying the met_em files from Ivan runs, replace the 4 soil moisture levels in them with the noFrances SM
#THIS CODE MODIFIES THE MET_EM FILE FOR DOMAIN 1 , other codes do the other 2 domains
import netCDF4 as nc
import os
import glob
import subprocess
import numpy as np
import matplotlib.pyplot as plt
import math
import xarray as xr
```

```python
file_paths1 = "/scratch/amanda/ivan_hires_ctrl_3dom_30s_terr/met_em.d01.2004-09-13_12:00:00.nc" #ivan
file_paths2 = '/scratch/amanda/metgrids_longer_studyperiod_preFrances_sm_ivan/met_em.d01.2004-09-06_06:00:00.nc'#dry
new_file_path = '/scratch/amanda/fixed_modified_met_em_noFrances_SM/' #putting edited met_em file into this dir
#for the "fixed" one, I modified the 3D SM variable as well in the met_em file, I only modified the 4 2D variables originally, and that did not work in the wrfoutput files
```

```python
with xr.open_dataset(file_paths1) as ds1:
    with xr.open_dataset(file_paths2) as ds2:
        file_name = file_paths1[45:] #just taking the last part of that filename (met_em part)
        ds1['SM000007'].values = ds2['SM000007'].values
        ds1['SM007028'].values = ds2['SM007028'].values
        ds1['SM028100'].values = ds2['SM028100'].values
        ds1['SM100289'].values = ds2['SM100289'].values

        # Replace values in SM for ds1 with values from ds2
       # loop over time dimension (needed to do this since the xarray dataset added a time dimension)
        for t in range(ds1.dims['Time']):
            # loop over soil layer dimension
            for layer in range(ds1.dims['num_sm_layers']):
                # assign values from ds2 to ds1
                ds1['SM'][t, layer, :, :] = ds2['SM'][t,layer, :, :]
          
        
        ds1.to_netcdf(path=new_file_path + file_name, mode='w', format="NETCDF4")
        print(type(ds1))
```
